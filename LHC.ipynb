```python
cnt = []
count = 0;
for i in tags:
  j = i.replace('><', ' ')
  ls = list(j.split())
  cnt.append(len(ls))

print(max(cnt))
print(min(cnt))
X = [1,2,3,4,5,6]
Y = []
for i in X:
  Y.append(cnt.count(i))
print(Y)
```

```python
cnt = []
count = 0;
unique = []
for i in tags:
  i=i.replace('>',' ')
  #print(i)
  i=i.replace('<', '')
  #print(i)
  ls = list(i.split())
  for j in ls:
    if j not in unique:
      unique.append(j)
print(len(unique))
```

```python
#Question 3 Top 25 tags appearing frequently
print(unique)
```

```python
count = 0;
countoftag=[0 for i in unique]
for i in tags:
  i=i.replace('>',' ')
  i=i.replace('<', '')
  ls = list(i.split())
  for j in ls:
    j = j.lower()
    index = unique.index(j)
    countoftag[index]+=1
print(countoftag)
    

#print(len(unique))
```
